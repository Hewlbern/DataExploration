#### Long story short - VK/VO according to our traffic scientistis is a decent indicator for obstruction, especially when compared with DS.

# Retrieve Data

###### Data comes from loop detectors - there are a variety of problems with loop detectors which influences the way we process the data. This includes just general faults, over-counting of cars, and under-counting. Loop detectors are induction loops in the ground that detect the passage of cars passing over them.

```python
#Create/Retireve SM Data for All regions on a specific Date

start_tot = time.time()
#dataframe = sm.import_sm_to_df('All', '20190717')
dataframe = pd.read_hdf('20190717_SA.h5')
end_tot = time.time()
print("Completed in " + str(end_tot-start_tot))
```

#### The data shown below shows Cycle Length of a intersection, Degree of Saturation, Reconstituted Volumes and Current Volumes.

### Degree of Saturation is arrival flow/ capacity. Capacity is maximum number of vehicles that can pass a given point. More mathematics but thats the principle.
### vk is the volume being picked up by the detector but with some weight average function. VK is therefore being calculated to ensure that the traffic volumes being used be SCATS are equivalent to “normal” traffic volumes. VK is usually the same as the actual volume collected by the detector. However it can be higher if: • there are slower moving vehicles due to congestion, or • long vehicles such as buss

### vo is just volumes.

```python
#Function that determines statistics for a days worth of data
def get_stats(group):
    diff = group.astype(int).abs().diff()
    ds_80 = (diff>80).sum()

    return {'min': group.min(), 'max': group.max(), 'count': group.count(),\
            'mean': group.mean(),'median': group.median(),'ds_80':ds_80,\
           'var': group.var(),'std': group.std()}
```

### This function creates congestion statistics for each Detector


```python
def get_VKVO_DS(selected_data):
    
    """
    Congestion stats for each site, selected data can be either "DS" or "vkvo ratio".
    """
    
    start_tot = time.time()
    detector_stats_dict = data[selected_data].groupby(['Region','Site','SA','Detector']).agg(get_stats)
    detector_stats = detector_stats_dict.reset_index().join(pd.DataFrame.from_records(detector_stats_dict.reset_index()[selected_data]))\
                            .drop(columns=[selected_data]).set_index(['Region','SA','Detector'])
    end_tot = time.time()
    print("Completed in " + str(end_tot-start_tot))
   

    filtered_dets = detector_stats.reset_index().set_index(['Region','SA'])

    print("Completed in " + str(end_tot-start_tot))
    #Sort by Highest daily mean 
    filtered_dets = filtered_dets.reset_index().sort_values(by=['mean'],ascending=False).set_index(['Region','Site','SA','Detector'])

    #Vk/Vo greater than 1
    filtered_dets = filtered_dets[filtered_dets['mean']>1]
    filtered_dets.head()
        #Add Sites names to Dataframe for readability from RAI Database csv 
    site_name_lookup = pd.read_excel('scats_computers.xlsx')
    site_name_lookup = site_name_lookup[['Site','Site Name']].set_index('Site')
    result = pd.merge(filtered_dets.reset_index(), site_name_lookup, on="Site")
    result = result.sort_values(by=['mean'],ascending=False).reset_index().set_index(['Site Name']).drop(columns = ['index','count','ds_80','max','median','min'])
    #Add Latitude and Longitude
    coords = pd.read_excel('intersection_coords.xlsx')
    coords = coords[['SITE_NO', 'LATITUDE','LONGITUDE']]
    coords = coords.rename(columns={"SITE_NO": "Site",}).set_index(['Site'])
    high_ds_dets = pd.merge(result.reset_index(),coords, on="Site")
    high_ds_dets = high_ds_dets.reset_index().sort_values(by=['mean'],ascending=False).set_index(['Region','Site','SA','Detector'])
    
    test = high_ds_dets.reset_index().set_index(['Site Name','Site','SA','Detector']).drop(columns=['index','Region','var'])
    test2 = test.reset_index().set_index("Site")
    print("Completed in " + str(end_tot-start_tot))
    return test2
```
